```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('LFP_26650_2Ah_0.1C_Chg_DChg.xlsx',sheet_name='record')
```

```python
print(df['Total Time'].dtype)
```

```python
print(df.columns)
```

```python
df = df.drop(columns='DataPoint')
```

```python


# Assuming df is your DataFrame
# Get descriptive statistics for all numeric columns
stats = df.describe()

# The 'min' and 'max' rows give you the range for each column
print(stats.loc[['min', 'max']])
```

```python
# Select only the numeric columns to calculate the correlation matrix
numeric_df = df.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
